```python
"""Print concentric polygons with increasing number of sides.
"""


import numpy as np

import vsketch

vsk = vsketch.Vsketch()

phase = -np.pi / 2
for i in range(20):
    angles = np.linspace(0, 2 * np.pi, i + 4)
    vsk.polygon((i + 1) * np.cos(angles + phase), (i + 1) * np.sin(angles + phase))

vsk.plot()
```
