# Polygons

Concentric polygons with increasing number of sides.

```python
%load_ext autoreload
%autoreload 2

#  use `widget` for interactive plot
%matplotlib inline


from matplotlib import rcParams

rcParams["figure.figsize"] = (14, 10)
```

```python
import numpy as np

import vsketch

vsk = vsketch.Vsketch()
vsk.scale("4mm")

phase = -np.pi / 2
for i in range(20):
    angles = np.linspace(0, 2 * np.pi, i + 4)
    vsk.polygon((i + 1) * np.cos(angles + phase), (i + 1) * np.sin(angles + phase))

vsk.plot()
vsk.write("polygons.svg", "a4")
```
