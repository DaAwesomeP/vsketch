# Shapely

[Shapely](https://shapely.readthedocs.io/en/latest/manual.html) is a powerful computational geometry often useful for generative art. Vsketch directly supports Shapely objects with the `Vsketch.geometry()` function.

```python
%load_ext autoreload
%autoreload 2

#  use `widget` for interactive plot
%matplotlib inline


from matplotlib import rcParams

rcParams["figure.figsize"] = (14, 10)
```

```python
import numpy as np
from shapely.geometry import Point
from shapely.ops import unary_union

import vsketch

vsk = vsketch.Vsketch()

shape = unary_union([Point(*np.random.random(2) * 5).buffer(np.random.random()) for _ in range(15)])
vsk.geometry(shape)
vsk.plot()
```
